```python
%matplotlib inline
import numpy
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import seaborn

seaborn.set(style='white')

patient = [98.9, 31, 25.4] 
cp = seaborn.color_palette('Blues_r', n_colors=3)
fig, ax = plt.subplots()
for n, p in enumerate(patient):
    circle = plt.Circle((1, p/2), radius=p/2, facecolor=cp[n])
    ax.add_artist(circle)
    ax.text(1, p / 2, str(p), color='black', fontsize=12, ha='center', va='center')

ax.set_xlim(-max(patient), max(patient))
ax.set_ylim(0, 1.2 * max(patient))
ax.set_aspect('equal')

plt.show()
```
