```python
import pandas as pd
import numpy as np
%matplotlib inline
import numpy
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import data

```python
df_authors = pd.read_excel('./data/input.xlsx',sheet_name="Feuil1")
df_authors.fillna(0, inplace=True)
```

```python
df_authors.columns = ['author_name', 'value3', 'value2', 'value1']
```

```python
df_authors = df_authors[['author_name', 'value1', 'value2', 'value3']]
```

```python
seaborn.set(style='white')

author = [98.9, 31, 25.4] 
cp = seaborn.color_palette('Blues_r', n_colors=3)
fig, ax = plt.subplots()
for n, p in enumerate(author):
    circle = plt.Circle((1, p/2), radius=p/2, facecolor=cp[n])
    ax.add_artist(circle)
    ax.text(1, p / 2, str(p), color='black', fontsize=12, ha='center', va='center')

ax.set_xlim(-max(author), max(author))
ax.set_ylim(0, 1.2 * max(author))
ax.set_aspect('equal')

plt.show()
```

```python
seaborn.set(style='white')

# authors data
author = [98.9, 31, 25.4]
author_name = "John Doe"  # Example author name
cp = seaborn.color_palette('Blues_r', n_colors=3)

# Create plot
fig, ax = plt.subplots()

# Plot each circle with the value displayed inside
for n, p in enumerate(author):
    circle = plt.Circle((1, p / 2), radius=p / 2, facecolor=cp[n], edgecolor='black', lw=1)
    ax.add_artist(circle)
    ax.text(1, p-5, str(p), color='black', fontsize=12, ha='left', va='top')

# Set x-axis to display author's name
ax.set_xlim(-max(author), max(author))
ax.set_ylim(0, 100)
ax.set_aspect('equal')

# Remove chart border
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.xaxis.set_ticks([])
plt.title(author_name)
plt.show()
```

```python
seaborn.set(style='white')
# Function to create a bubble chart for a group of authorss
def create_bubble_chart(df_group, fig_num):
    fig, axes = plt.subplots(4, 4, figsize=(20, 20))
    axes = axes.flatten()
    
    colors = ['green', 'red', 'grey']  # Define three distinct colors
    
    for i, (index, row) in enumerate(df_group.iterrows()):
        ax = axes[i]

        for n, value in enumerate([row['value1'], row['value2'], row['value3']]):
            if value > 0:
                circle = plt.Circle((1, value / 2), radius=value / 2, facecolor=colors[n], edgecolor='white')
                ax.add_artist(circle)
                ax.text(1, value-20, f"{value:.1f}", ha='center', va='center', color='white', fontsize=12)

        ax.set_xlim(-max_author_value / 1.8, max_author_value / 1.8)
        ax.set_ylim(0, max_author_value)
        ax.set_aspect('equal')
        # Remove chart border
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.yaxis.set_ticks([])
        ax.xaxis.set_ticks([])
        ax.set_title(row['author_name'],y=-0.1, ha='center')

    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
#     plt.suptitle(f'Chart {fig_num}', fontsize=20)
    plt.subplots_adjust(top=0.95)
    plt.show()

# Find the global max value for xlim and ylim
max_author_value = df_authors[['value1', 'value2', 'value3']].values.max()

# Split the authors into groups of 4 and create charts
for i in range(0, len(df_authors), 4):
    df_group = df_authors.iloc[i:i+4]
    create_bubble_chart(df_group, fig_num=(i // 4 + 1))
```
